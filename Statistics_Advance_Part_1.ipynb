Question 1: What is a random variable in probability theory?

A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each outcome in the sample space of a probabilistic experiment. There are two types: discrete and continuous.

Question 2: What are the types of random variables?

The two main types of random variables are:

Discrete Random Variable: Takes a countable number of distinct values.

Continuous Random Variable: Takes an infinite number of possible values within a given range.

Question 3: What is the difference between discrete and continuous distributions?

Discrete distributions deal with countable outcomes (e.g., number of heads in coin tosses), while continuous distributions deal with outcomes over a continuum (e.g., height, weight).

Question 4: What are probability distribution functions (PDF)?

A Probability Distribution Function (PDF) describes the likelihood of a random variable taking on a specific value. For discrete variables, it's a probability mass function (PMF); for continuous, it's a function where the area under the curve gives the probability.

Question 5: How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

CDF gives the probability that a random variable is less than or equal to a particular value, while PDF describes the relative likelihood for the variable to take on a specific value.

Question 6: What is a discrete uniform distribution?

A discrete uniform distribution is one where all outcomes are equally likely. For example, rolling a fair die has outcomes 1 through 6, each with equal probability 1/6.

Question 7: What are the key properties of a Bernoulli distribution?

Only two possible outcomes: success (1) and failure (0)

One trial

Probability of success is p, failure is (1-p)

Mean = p, Variance = p(1-p)

Question 8: What is the binomial distribution, and how is it used in probability?

A binomial distribution represents the number of successes in a fixed number of independent Bernoulli trials. It is used in scenarios like the number of heads in multiple coin tosses.

Question 9: What is the Poisson distribution and where is it applied?

The Poisson distribution models the number of times an event occurs in a fixed interval of time or space. It's used in areas such as call center traffic or decay events per unit time.

Question 10: What is a continuous uniform distribution?

In a continuous uniform distribution, all outcomes in a continuous range are equally likely. The PDF is constant over the interval.

Question 12: What is the standard normal distribution, and why is it important?

A standard normal distribution has a mean of 0 and a standard deviation of 1. It's important because many statistical tests are based on it and it simplifies calculations.

Question 13: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution. It justifies the use of normal distribution in inferential statistics.

Question 14: How does the Central Limit Theorem relate to the normal distribution?

The CLT explains why many distributions tend to be normal in large samples, making normal-based methods widely applicable in statistics.

Question 15: What is the application of Z statistics in hypothesis testing?

Z statistics help determine how far a sample mean is from the population mean in units of standard deviation. It's used to test hypotheses and calculate p-values.

Question 16: How do you calculate a Z-score, and what does it represent?

Z = (X - μ) / σ
It represents how many standard deviations a value X is from the mean μ.

Question 17: What are point estimates and interval estimates in statistics?

Point Estimate: A single value estimate of a population parameter (e.g., sample mean)

Interval Estimate: A range of values within which the parameter is expected to lie (e.g., confidence interval)

Question 18: What is the significance of confidence intervals in statistical analysis?

Confidence intervals provide a range of plausible values for the population parameter, offering a measure of precision and reliability of estimates.

Question 19: What is the relationship between a Z-score and a confidence interval?

Z-scores define the margin of error in confidence intervals. For example, a 95% confidence interval uses a Z-score of approximately 1.96.

Question 20: How are Z-scores used to compare different distributions?

Z-scores standardize values from different distributions, allowing direct comparison even if the distributions have different means or standard deviations.

Independent random samples

Question 22: What is the concept of expected value in a probability distribution?

Expected value is the long-term average or mean of a random variable. For discrete variables: E(X) = Σ[x * P(x)]

Question 23: How does a probability distribution relate to the expected outcome of a random variable?

The probability distribution determines the weights (probabilities) of each outcome, directly affecting the calculation of the expected value.
